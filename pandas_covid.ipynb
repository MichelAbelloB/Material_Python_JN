```python
data_cov['date'] = pd.to_datetime(data_cov['date'])
print(data_cov['date'])
```

```python
data_cov = data_cov.set_index('date')
```

```python
display(data_cov)
```

```python
data_cov_col = data_cov[data_cov['location'] == 'Colombia']
plt.figure(figsize = (10,3))
plt.plot(data_cov_col.index, data_cov_col['total_deaths'])
plt.show()
```

```python
plt.figure(figsize = (10,3))
plt.plot(data_cov_col.index, data_cov_col['total_cases'])
plt.show()
```

```python
data_cov_sa = data_cov[data_cov['continent'] == 'South America']
display(data_cov_sa)
```

```python
data_cov_sa_total_cases = data_cov_sa[['iso_code','total_cases']].reset_index()
data_cov_sa_total_cases['total_cases'] = np.log(data_cov_sa_total_cases['total_cases'])
print(data_cov_sa_total_cases.head())
data_cov_sa_total_cases = data_cov_sa_total_cases.pivot(index='date', columns='iso_code', values='total_cases')
print(data_cov_sa_total_cases.tail())
data_cov_sa_total_cases.plot(figsize = (15,10), title = 'Total casos SudamÃ©rica',
                             xlabel = 'Fecha', ylabel = 'Total casos')
plt.savefig('tc_sudA.pdf', format = 'pdf')
plt.show()
```

```python
data_cov_sa_total_deaths = data_cov_sa[['iso_code','total_deaths']].reset_index()
data_cov_sa_total_deaths['total_deaths'] = np.log(data_cov_sa_total_deaths['total_deaths'])
print(data_cov_sa_total_deaths.head())
data_cov_sa_total_deaths = data_cov_sa_total_deaths.pivot(index='date', columns='iso_code', values='total_deaths')
print(data_cov_sa_total_deaths.tail())
data_cov_sa_total_deaths.plot(figsize = (15,10), title = 'Total muertes SudamÃ©rica',
                             xlabel = 'Fecha', ylabel = 'Total muertes')
plt.savefig('td_sudA.pdf', format = 'pdf')
plt.show()
```

```python
data_cov_acc = data_cov[data_cov.index == datetime.datetime(2021, 6, 7)].dropna(subset= ['continent'])
print(data_cov_acc)
```

```python
import seaborn as sns
plt.figure(figsize = (10,5))
sns.scatterplot(x = 'gdp_per_capita', y = 'total_deaths_per_million', 
                hue = 'continent',  size = 'population_density', 
                data = data_cov_acc)
plt.savefig('scatter_data.pdf', format = 'pdf')
plt.show()
```
