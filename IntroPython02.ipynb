# Funciones

- Útiles para tener un código compacto y evitar repetir código
- Corren únicamente cuando se llaman.


```python
def square(x):
    return x*x
print(square(5))
print(square(-3))
```

```python
def user(name, id, age, gender = 'F'):
    print('Creando usuario ...')
    print('Nombre: {} | \n CC: {} | \n Edad: {} | \n Género: {}'.format(name, id, age, gender))
    print('Usuario creado con éxito.')

user('Juan', 723254, 50, 'M')
user('Beatriz', 465158, 35)
```

### Posición de los argumentos

```python
def create_email(nombre1, nombre2, apellido1, apellido2):
    return nombre1[0].lower() + nombre2[0].lower() + apellido1.lower() + apellido2[0].lower() + '@correo.edu.co'
email0 = create_email(apellido1 = 'Reyes', nombre1 = 'Ricardo', apellido2 = 'Basoalto', nombre2 = 'Eliecer') #Nótese el orden
print(email0)
```

## Funciones anónimas: expresiones `lambda`
- Funciones en línea usadas para tareas pequeñas y una sola vez
- No es obligatorio que tengan un nombre, por eso se llaman también funciones anónimas.

```python
power = lambda x, n: x**n 
print(power(5,3))
```

### Aplicar función a un objeto iterable

### map()

```python
def count_let(a):
    return len(a)

list0 = ['cien', 'años', 'de', 'soledad']
x = list(map(count_let, list0))
print(x)

y = list(map(lambda s: len(s), list0))
print(y)
```

### filter()

```python
def greater_than_10(x):
    return x > 10


list1 = [30, 5, 12, 1.02, 0]
x = list(filter(greater_than_10, list1))
print(x)


y = list(filter(lambda s: s > 10, list1))
print(y)
```
